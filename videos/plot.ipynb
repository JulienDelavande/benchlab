```python
FOLDER = '/fsx/jdelavande/benchlab/videos/data'
CSV1 = 'emissions_base_5a55b494-af9a-46be-8ab0-9d81ab8ae29e.csv'
CSV2 = 'emissions_base_6d00eae0-a8d7-4ad9-9037-e086fd966d68.csv'
CSV3 = 'emissions_base_7bfcb595-770f-46f5-9536-2d59e441acf9.csv'
CSV4 = 'emissions_base_96ae8db5-25d2-479d-903e-89c54644db2c.csv'
CSV5 = 'emissions_base_a3b4b17c-ca96-411a-88f3-150a15de0af4.csv'


import pandas as pd
df1 = pd.read_csv(f'{FOLDER}/{CSV1}')
df2 = pd.read_csv(f'{FOLDER}/{CSV2}')
df3 = pd.read_csv(f'{FOLDER}/{CSV3}')
df4 = pd.read_csv(f'{FOLDER}/{CSV4}')
df5 = pd.read_csv(f'{FOLDER}/{CSV5}')
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Si ce n'est pas déjà fait, charge ton DataFrame ici
# df = pd.read_csv("ton_fichier.csv")

# Liste des types d'énergie
energy_types = ['gpu_energy', 'cpu_energy', 'ram_energy']
colors = ['#4E79A7', '#F28E2B', '#59A14F']

# Calcul des moyennes et écarts-types par tâche et par type d'énergie
grouped = df.groupby('task_name')[energy_types].agg(['mean', 'std'])

# Préparation du plot
tasks = grouped.index.tolist()
x = range(len(tasks))
bar_width = 0.25

plt.figure(figsize=(10, 6))

for i, energy in enumerate(energy_types):
    means = grouped[(energy, 'mean')]*1000
    stds = grouped[(energy, 'std')]*1000
    offsets = [pos + i * bar_width for pos in x]
    plt.bar(offsets, means, yerr=stds, width=bar_width, label=energy.replace('_energy', '').upper(), color=colors[i], capsize=5, alpha=0.7, edgecolor='black')

# Mise en forme
plt.xticks([pos + bar_width for pos in x], tasks)
plt.ylabel("Energy (Wh)")
plt.title("Energy Consumption of text to video generation tasks, Lightricks/LTX-Video-0.9.7-dev H100 GPU")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from brokenaxes import brokenaxes

# Liste des types d'énergie
energy_types = ['gpu_energy', 'cpu_energy', 'ram_energy']
colors = ['#4E79A7', '#F28E2B', '#59A14F']

# Calcul des moyennes et écarts-types par tâche et par type d'énergie
grouped = df.groupby('task_name')[energy_types].agg(['mean', 'std'])

# Préparation du plot avec coupure d'axe
tasks = grouped.index.tolist()
x = range(len(tasks))
bar_width = 0.25

fig = plt.figure(figsize=(10, 6))
bax = brokenaxes(ylims=((0, 0.25), (1.25, 2.0)), hspace=0.05)

for i, energy in enumerate(energy_types):
    means = grouped[(energy, 'mean')]*1000
    stds = grouped[(energy, 'std')]*1000
    offsets = [pos + i * bar_width for pos in x]
    bax.bar(offsets, means, yerr=stds, width=bar_width, label=energy.replace('_energy', '').upper(), 
            color=colors[i], capsize=5, alpha=0.7, edgecolor='black')

# Mise en forme
bax.set_xticks([pos + bar_width for pos in x])
bax.set_xticklabels(tasks)
bax.set_ylabel("Energy (Wh)")
bax.set_title("Energy Consumption of text to video generation tasks, Lightricks/LTX-Video-0.9.7-dev H100 GPU")
bax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
